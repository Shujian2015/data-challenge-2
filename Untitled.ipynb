```python
%matplotlib inline  
import warnings
warnings.filterwarnings('ignore')

from datetime import datetime, timedelta
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import math
import pandas as pd
import seaborn as sns
sns.set(style="ticks")

import sklearn.metrics
from matplotlib import pyplot as plt
from sklearn.metrics import classification_report
```

### Load the datasets and join

```python
test = pd.read_csv('test_results.csv')
test.head()
```

```python
users = pd.read_csv('user_table.csv')
users.head()
```

```python
df = test.join(users.set_index('user_id'), on='user_id')
df.head()
```

```python
df.describe()
```

```python
def change_time(time_str):
    time_str = time_str[0:10]
    datetime_object = datetime.strptime(time_str, '%Y-%m-%d')
    return datetime_object


df['time'] = df['timestamp'].apply(lambda x: change_time(x))
first = df['time'].min()
df['days'] = df['time'].apply(lambda x: (x - first).days)
df.head()
```

### Clean data

```python
cols_to_transform = ['source', 'device', 'operative_system']
df_with_dummies = pd.get_dummies(df, columns = cols_to_transform)

cols_to_drop = ['user_id', 'timestamp', 'price', 'city', 'country', 'time']
df_with_dummies.drop(cols_to_drop, axis=1, inplace=True)

df_with_dummies.fillna(df_with_dummies.mean(), inplace=True)

df_with_dummies.head()
```

```python
df_with_dummies.describe()
```

```python
columns_to_display = ['days', 'test', 'lat', 'long', 'converted']
sns.pairplot(df_with_dummies[columns_to_display], hue="converted")
# sns.pairplot(df[])
```

### Machine learning with RF

```python
x = df_with_dummies.copy().drop(['converted'], axis=1)
y = df_with_dummies['converted']

x_train, x_valid, y_train, y_valid = train_test_split(x, y, test_size=0.3, random_state=2017)

clf = RandomForestClassifier()
model = clf.fit(x_train, y_train)

y_pred = clf.predict(x_valid)

print(classification_report(y_valid, y_pred, target_names=['0', '1']))

# http://www.markhneedham.com/blog/2017/06/16/scikit-learn-random-forests-feature-importance/
headers = ["name", "score"]
values = sorted(zip(x.columns, model.feature_importances_), key=lambda x: x[1] * -1)

print('')
print('Feature Importance :')
for value in values:
    print(value)
```

#### Discussion:

As shown in the above feature importance list, test is not a strong feature in this case. It can assume that company can sell its software $59.
